```python
from selenium import webdriver
```

```python
browser = webdriver.Chrome('./chromedriver')
browser.get('https://github.com/login')
```

```python

# id  input#login_field
# pwd  input#password
# btn input[type="submit"]
```

```python
import time

browser.find_element_by_css_selector('input#login_field').send_keys('rcm2000')
browser.find_element_by_css_selector('input#password').send_keys('')
browser.find_element_by_css_selector('input[type="submit"]').click()
time.sleep(2)
browser.find_element_by_css_selector('a[href="/rcm2000/test_webscraping"]').click()
```

```python
html = browser.page_source
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html,'html.parser')
```

```python
titles = soup.select('div[role="row"] > div[role="rowheader"]')
times = soup.select('div[role="row"] time-ago.no-wrap')
commits = soup.select('div[role="row"] a[data-pjax="true"]') 
commits
```

```python
a = []
for i in range(0,len(times)):
    title = titles[i].text.strip()
    time = times[i].text.strip()
    datetime = times[i]['title'].strip('GMT+9')
    commit = commits[i].text.strip()
    a.append([title,commit,datetime,time])
a
   
```

```python
import pandas as pd
```

```python
df1 = pd.DataFrame(a,columns = ['Title','commit','datetime','time'])
```
