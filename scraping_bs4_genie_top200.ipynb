```python
import requests
```

```python
url = 'https://music.bugs.co.kr/chart'
req = requests.get(url)
```

```python
from bs4 import BeautifulSoup as bs
```

```python
soup = bs(req.content,'html.parser')
```

```python
soup
```

```python
li = soup.select('p.title > a')
```

```python
len(li)
```

```python
list = []
for i in li:
    list.append(i.text)
```
