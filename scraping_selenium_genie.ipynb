```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome('./chromedriver.exe')
```

```python
url = 'https://www.genie.co.kr/chart/top200'
```

```python
driver.get(url)
```

```python
html = driver.page_source
```

```python
from bs4 import BeautifulSoup as bs
```

```python
soup = bs(html,'html.parser')
```

```python
tags = soup.select('.title.ellipsis')
```

```python
for i in tags:
    print(i.text.strip())
```

```python
artist = soup.select('.info > .artist.ellipsis')
```

```python
for j in artist:
   print(j.text)
    
```

```python
li = []
for k in range(0,len(tags)):
    li.append([tags[k].text.strip(),artist[k].text.strip()])
```

```python
li
```

```python
import pandas as pd
```

```python
df1 = pd.DataFrame(li , columns = ['TItle','Artist'])
```

```python
b = []
for rank in range(1,51):
    b.append(str(rank)+'위')
```

```python
df1['rank'] = b
```

```python
df2 = df1.set_index('rank')
```
